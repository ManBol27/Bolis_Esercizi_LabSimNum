## Pricing delle opzioni PUT e CALL europee

### Cenni teorici
Le opzioni "plain vanilla" europee sono strumenti finanziari che consistono in un contratto tra due parti in cui la prima ha la possibilità ma non l'obbligo di acquistare (CALL) oppure vendere (PUT) un determinato asset (il sottostante), ad un certo prezzo $K$ ("strike price") in una certa data ("delivery time"), mentre la controparte è obbligata ad eseguire/non eseguire lo scambio a secondo del fatto che venga o meno esercitata l'opzione.

Il sottostante può essere un'azione, un indice o altri strumenti finanziari, ai quali l'opzione è legata. Per natura le opzioni appartengono dunque alla categoria degli strumenti finanziari cosiddetti "derivati".
Colui che ha acquistato l'opzione eserciterà il suo diritto nel caso in cui al delivery time T il prezzo del sottostante S(T) è: 
* $>\; K$ (per opzione CALL)
* $<\; K$ (per opzione PUT).
Realizzando un **profitto** rispettivamente pari a **$S(T) - K$ per opzione CALL e a $K - S(T)$ per opzione PUT**, al quale va scontato il prezzo pagato per acquistare l'opzione.

### Esempio pratico
Un investitore acquista un'opzione CALL ad un prezzo di $30\$$ il cui sottostante sono le azioni TESLA, dal valore attuale di mercato pari a $280\$$. Lo strike price $K$ è fissato allo stesso prezzo di $280\$$ e il delivery time $T$ è a 30 giorni dall'acquisto dell'opzione.

**Scenario 1** (favorevole): 
Al delivery time le azioni TESLA hanno aumentato il loro valore ed ora hanno un prezzo di $330\$$, l'investitore esercita l'opzione di acquistare le azioni ad un prezzo pari allo strike price ($280\$$) e le rivende immediatamente sul mercato al prezzo attuale, realizzando così un profitto al netto del prezzo inizialmente pagato per l'opzione pari a $330-280-30 = 20\$$. Dunque l'investitore ha guadagnato $20\$$ avendone investiti $30\$$ (ha pagato soltanto il prezzo dell'opzione!), realizzando così un profitto del 67\% circa a fronte di una rivalutazione del sottostante del 18\%.

**Scenario 2** (sfavorevole): 
Al delivery time le azioni TESLA hanno subìto anche soltanto una lieve correzione ed ora prezzano $270\$$ sul mercato, l'investitore non esercita il suo diritto di acquistare il sottostante e l'opzione scade senza essere esercitata. In questo caso l'investitore ha perso il 100\% del capitale investito (ovvero il prezzo dell'opzione pagato inizialemente).

In questo caso (dove lo strike è fissato ad un valore pari al prezzo attuale) un investimento in opzioni può portare ad un profitto con una certa leva rispetto ad acquistare direttamente l'azione sottostante, ma nel caso sfavorevole di una pur piccola correzione può portare alla perdita totale del capitale investito.

Nota: lo strike non è per forza uguale al prezzo totale, esistono infatti diverse strategie di investimento che utilizzano opzioni "in the money" (se strike < prezzo attuale del sottostante) piuttosto che opzioni "out of the money" (se strike > prezzo attuale del sottostante).

### Pricing di un'opzione
Il prezzaggio delle opzioni, ovvero l'operazione di stabilire quale sia il "giusto" prezzo di una opzione, segue da due concetti fondamentali, ove il secondo discende in qualche modo dal primo:
* Non arbitraggio: il "giusto prezzo" di un’opzione è quello al quale non esistono opportunità di arbitraggio, ossia situazioni in cui un investitore potrebbe ottenere un profitto senza rischio.
* Rischio neutro: in questo quadro, tutti gli investimenti rendono il tasso privo di rischio ("risk free interest rate"), ovvero quello offerto dalle banche in un dato momento sul deposito di liquidità.


Come accennato, si può dimostrare con degli esempi che se venisse utilizzato il rendimento atteso dal sottostante anzichè il tasso risk free per prezzare un'opzione si potrebbero creare opportunità di arbitraggio.

Il valore attuale di un'opzione è calcolato come il payoff (rispettivamente pari a $S(T) - K$ per opzione CALL e a $K - S(T)$ per opzione PUT) scontato dell'interesse che offrirebbe un invesitmento risk free, infatti: 

Se il valore attuale è $e^{-rT} \max\{0 \; , \pm (S(T)-K) \}$  ($+$ per opzioni CALL, $-$ per opzioni PUT)

Allora il valore finale (profitto) sarebbe $e^{rT} \; e^{-rT} \max\{0 \; , \pm (S(T)-K) \} = \max\{0 \; , \pm (S(T)-K) \}$

come ci si aspetta.

Il senso di quanto appena scritto è:  

* Se si investe l’importo attuale (il prezzo dell’opzione) al tasso risk free, si otterrà al tempo T esattamente il payoff dell’opzione, in assenza di arbitraggio.

* Questo è il principio alla base del pricing risk neutral: il portafoglio replicante (che è privo di rischio) cresce al tasso risk free, quindi il prezzo attuale deve essere tale che, investito a questo tasso, dia il payoff finale dell’opzione.

### Simulazione dell'andamento del sottostante
Resta ora da capire come simulare il valore del sottostante $S(T)$. Si può considerare l'evoluzione di $S(t)$ come un moto browniano geometrico GBM (aka un moto browniano con drift, in altre parole un'andamento medio prevedibile sommato ad una componente del tutto stocastica). Per simulare questo tipo di andamento esistono due modi:
* Sampling diretto di $S(T)$ da una $GBM (\mu, \sigma^2)$ dove $\mu$ è il drift e $\sigma$ è la volatilità
* Sampling a tempi discreti del $GBM (\mu, \sigma^2)$ dividendo l'intervallo $[0,T]$ in un certo numero di sottointervalli e valutare GBM in ognuno di questi $t_i \in t_0, \; t_1,\; ... ,\; T$.


L'espressione per $S\sim GBM(\mu,\sigma^2)$, con valore iniziale $S(0)$ è:
$$
S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]
$$
dove $W(t)\sim\mathcal{N}(0,t)$. 

Inoltre, siccome gli incrementi di $W$ sono indipendenti e normalmente distribuiti, la procedura ricorsiva per estrarre i valori da $S\sim GBM(\mu,\sigma^2)$ a $t_1 < t_2 < … < t_n$ è:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$

### Media cumulativa con errore
Come svolto nelle esercitazioni precedenti, per dare una stima del valore atteso con relativo errore per il prezzo delle opzioni si sono eseguite un gran numero di simulazioni ($10^6$) per i valori di $S(t)$ per poi calcolare:
$$
C(t) = e^{-rT} \max\{0 \; , \pm (S(T)-K) \}
$$
ovvero il prezzo attuale dell'opzione (+ per azione CALL, - per azione PUT).

Successivamente si sono raggruppati questi valori in $N = 100$ blocchi con $L=10000$ simulazioni di $C(t)$ ciascuno per calcolare la media cumulativa con relativo errore.

### Modello Black-Scholes
Esiste inoltre un modello analitico con cui calcolare il prezzo di un opzione, che costituisce uno dei cardini della matematica finanziaria, sviluppato da Black e Scholes nel 1973, con cui verranno confrontati i risultati ottenuti con il metodo Monte Carlo descritto sopra. Si riporta il risultato per completezza:

Per il prezzo al a tempi t<T:
- Prezzo opzione call europea:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- Prezzo opzione put europea:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

dove $d_2 = d_1 - \sigma\sqrt{T-t}$, e
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
è la gaussiana cumulativa $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
e dove
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

### Parametri utilizzati
Il modello descritto è stato testato impostando i seguenti parametri finanziari: 
- Prezzo del sottostante al tempo $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatilità: $\sigma=0.25$

```python
import scipy
from scipy.special import erf
import numpy as np
import matplotlib.pyplot as plt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / np.sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * np.sqrt(T)) * (np.log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * np.sqrt(T)
    C = S0 * N(d1) - K * np.exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * np.exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

```python
dir_call, err_dir_call = np.loadtxt("CALL_diretto.txt", unpack= True, usecols=(0,1), skiprows=1)
dir_put, err_dir_put = np.loadtxt("PUT_diretto.txt", unpack= True, usecols=(0,1), skiprows=1)
dis_call, err_dis_call = np.loadtxt("CALL_discreto.txt", unpack= True, usecols=(0,1), skiprows=1)
dis_put, err_dis_put = np.loadtxt("PUT_discreto.txt", unpack= True, usecols=(0,1), skiprows=1)

N_blocks = len(dir_call) # numero blocchi

# x del grafico, numero di lanci totali incrementando ogni volta il numero di blocchi
x = np.arange(1, N_blocks+1)     # numero blocchi

plt.figure(figsize=(17,13))

##################################
# Grafico prezzo Call caso dicreto
plt.subplot(2,2,1)

plt.title(r"Prezzo opzione Call (diretto)", fontsize=18)

plt.ylim(14.45, 15.35)
plt.errorbar(x, dir_call, yerr=err_dir_call, label = "Call simulato", color = 'blue') #plotto la media cumulativa con errore del prezzo di Call

plt.xlabel(r"Blocchi", fontsize=18)
plt.ylabel(r"Prezzo opzione Call", fontsize=18)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.axhline(y=call, xmin=0, xmax=1, label=r'Call teorico= %.2f' %(call), lw = 2, color = 'red')

plt.legend(loc='lower right', prop={'size': 12})

plt.grid(True)

######################################
# Plot prezzo Put caso diretto

plt.subplot(2,2,3)

plt.title(r"Prezzo opzione Put (diretto)", fontsize=18)

plt.ylim(5.15, 5.85)
plt.errorbar(x, dir_put, yerr=err_dir_put, label = "Put simulato", color = 'blue') #plotto la media cumulativa con errore del prezzo di Call

plt.xlabel(r"Blocchi", fontsize=18)
plt.ylabel(r"Prezzo opzione Put", fontsize=18)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.axhline(y=put, xmin=0, xmax=1, label=r'Put teorico= %.2f' %(put), lw = 2, color = 'red')

plt.legend(loc='lower right', prop={'size': 12})

plt.grid(True)

#####################################
# Grafico prezzo Call caso discreto
plt.subplot(2,2,2)

plt.title(r"Prezzo opzione Call (discreto)", fontsize=18)

plt.ylim(14.45, 15.35)
plt.errorbar(x, dis_call, yerr=err_dis_call, label = "Call simulato", color = 'green') #plotto la media cumulativa con errore del prezzo di Call

plt.xlabel(r"Blocchi", fontsize=18)
plt.ylabel(r"Prezzo opzione Call", fontsize=18)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.axhline(y=call, xmin=0, xmax=1, label=r'Call teorico= %.2f' %(call), lw = 2, color = 'red')

plt.legend(loc='lower right', prop={'size': 12})

plt.grid(True)

######################################
# Plot prezzo Put caso discreto

plt.subplot(2,2,4)

plt.title(r"Prezzo opzione Put (discreto)", fontsize=18)

plt.ylim(5.15, 5.85)
plt.errorbar(x, dis_put, yerr=err_dis_put, label = "Put simulato", color = 'green') #plotto la media cumulativa con errore del prezzo di Call

plt.xlabel(r"Blocchi", fontsize=18)
plt.ylabel(r"Prezzo opzione Put", fontsize=18)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.axhline(y=put, xmin=0, xmax=1, label=r'Put teorico= %.2f' %(put), lw = 2, color = 'red')

plt.legend(loc='lower right', prop={'size': 12})

plt.grid(True)

plt.show()

print("Valor medio prezzo opzione Call (campionamento diretto): ", round(dir_call[-1],3), "+/-", round(err_dir_call[-1],3))
print("Livello di confidenza con il valore atteso: ",np.abs(round((dir_call[-1] - call)/err_dir_call[-1],2)), "sigma")
print("\n")

print("Valor medio prezzo opzione Call (campionamento discreto): ", round(dis_call[-1],3), "+-", round(err_dis_call[-1],3))
print("Livello di confidenza con il valore atteso: ",np.abs(round((dis_call[-1] - call)/err_dis_call[-1],2)), "sigma")
print("\n ---------------------")

print("Valor medio prezzo opzione Put (campionamento diretto): ", round(dir_put[-1],3), "+/-", round(err_dir_put[-1],4))
print("Livello di confidenza con il valore atteso: ", np.abs(round((dir_put[-1] - put)/err_dir_put[-1],2)), "sigma")
print("\n")

print("Valor medio prezzo opzione Put (campionamento discreto): ", round(dis_put[-1],3), "+/-", round(err_dis_put[-1],4))

print("Livello di confidenza con il valore atteso: ", np.abs(round((dis_put[-1] - put)/err_dis_put[-1],2)), "sigma")
```

I risultati non evidenziano particolari differenze tra i due metodi utilizzati. In tutti i casi l'accordo con il valore atteso (calcolato analiticamente con la formula di Black-Scholes) è buono o molto buono.
