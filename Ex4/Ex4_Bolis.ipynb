# Esercitazione 4 - Dinamica molecolare

### Simulazione di un sistema composto da molecole
L'algoritmo fornito simula la dinamica di un sistema a $N$ particelle (molecole) interagenti mediante un potenziale di Lennard-Jones:
$$
 V = 4 \epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] $$

dove $\epsilon$ è la profondità della buca di potenziale, $\sigma$ è il raggio della sfera che approssima l'atomo o la molecola in un modello a sfera rigida e $r_{i}-r_{j}$ sono le distanze tra coppie di particelle interagenti. Il carattere di questo potenziale è repulsivo per piccole distanze intermolecolari e attrattivo quando le distanze intermolecolari aumentano.

Seguendo la dinamica vincolata ad energia costante, il codice è in grado di tenere traccia dei valori di energia cinetica, energia potenziale, energia totale, temperatura e pressione. Quest'ultima calcolata usando il viriale:
$$
 P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle 
$$
dove $\epsilon$, $\sigma$, $r_{i}-r_{j}$ definite come sopra, $V$ è il volume occupato dalle particelle, $T$ è la temperatura del sistema, $\rho$ è la densità di particelle.

Per questo problema è opportuno lavorare in unità ridotte di Lennard-Jones, rispettivamente $r^* = \frac{r}{\sigma}$ come unità di lunghezza, $E^* = \frac{E}{\epsilon}$ come unità di energia e $T^* =\frac{T}{\epsilon / k_B}$ come unità di temperatura. Questa scelta permette di descrivere più sistemi con diverse combinazioni di temperatura, energia e distanza intermolecolare che rispecchiano però le stesse unità ridotte, normalizzando ciascun sistema con rispettive $\sigma$, $\epsilon$ e $\epsilon / k_B$. 
Spieghiamo ora più in dettaglio l'algoritmo di Verlet.

### Algoritmo di Verlet 
Una volta selezionato un potenziale, è possibile da esso dedurre la forza e fare evolvere il sistema con le equazioni della dinamica classica. La procedura consiste nell'integrare il sistema per piccoli passi temporali tenendo traccia dell'evoluzione dello spazio delle fasi (aka delle posizioni e dei momenti delle $N$ particelle che compongono il sistema. Si noti come anche il potenziale, dipendendo dalle posizioni, dipende a sua volta dal tempo. Dunque, come riportato nel paragrafo precedente, la conoscenza dell'evoluzione di queste variabili permette di ricavare l'evoluzione di grandezze quali l'energia totale per particella $E/N$, l'energia cinetica per particella $K/N$, l'energia potenziale per particella $U/N$, temperatura $T$ e pressione $P$ del sistema.  
L'algoritmo di Verlet consiste in quanto segue:  

si parte da un'espansione di Taylor al terzo ordine della coordinata $r$ (posizione), per ogni particella. In particolare si fa uno sviluppo $r(t+\delta t)$ e di $r(t-\delta t)$, e sommando i due sviluppi si ottiene
$$
\vec{r}(t+\delta t) = 2 \vec{r}(t) - \vec{r}(t-\delta t) + \delta t^2 \cdot\vec{a}(t) + o(\delta t^4)
$$
da cui si osserva che quello appena descritto è un metodo "two-steps", infatti la posizione al tempo $t+\delta t$ dipende dalla posizione al tempo $t$ e al tempo $t-\delta t$. Si vede anche che non c'è dipendenza diretta dalla velocità, ma soltanto dall'accelerazione, quest'utlima ricavata dal valore del potenziale ad un certo step. Ovviamente si può ricavare la velocità conoscendo la posizione in istanti successivi di tempo, che è ciò che fa l'algoritmo per poi calcolare l'energia cinetica e la temperatura.

### Periodic boundary conditions 
Lo studio di un sistema confinato entro un certo volume diventa complicato quando si devono considerare le molecole vicino ai bordi, le quali sentono forze diverse rispetto alle altre molecole. Un modo per modellare più facilmente il sistema è quello di suddividere quest'ultimo in un reticolo composto da volumetti cubici che si ripetono periodicamente all'infinito. Questo approccio semplifica notevolmente anche il costo computazionale in termini di memoria, in quanto sarà sufficente conoscere lo spazio delle fasi per le molecole dentro ad un singolo cubo per poi trattare le altre molecole come "immagini" di quest'ultime. E' sufficiente quindi conoscere il numero $N$ di particelle e la densità $\rho$ per ricavare il volume di un singolo cubo del reticolo. 
Sempre nell'ottica di ridurre il costo computazionale dell'algoritmo una scelta comune è quella di prendere un raggio di cutoff $r_c$ attorno ad certa una molecola e considerare soltanto le interazioni tra quella molecola e le molecole contenute nella sfera di raggio $r_c$. E' opportuno aumentare il raggio di cutoff quando le molecole interagiscono poco, ad esempio nel caso in cui consideri un gas, mentre si può utilizzare un $r_c$ minore nel caso di un solido, dove ciascuna molecola interagisce in maniera significativa soltanto con le molecole più vicine nel reticolo cristallino.

### Temperatura di equilibrio
La simulazione viene avviata impostando come configurazione iniziale per le molecole un reticolo cristallino, la quale minimizza il potenziale di Lennard-Jones. Impostando una certa temperatura iniziale per la simulazione, dalla quale vengono settate le velocità iniziali delle singole molecole estraendole da una distribuzione di Maxwell-Boltzmann, si può notare dalla simulazione come essa non venga mantenuta nel tempo, ma tenda invece a diminuire per poi assestarsi ad un valore inferiore. Questo accade poichè le molecole aventi una certa velocità iniziale tenderanno  ad avvicinarsi tra di loro aumentando così l'energia potenziale (carattere repulsivo in Lennard-Jones), dovendosi poi conservare l'energia totale il risultato è una diminuzione dell'energia cinetica (e quindi della temperatura) fino al raggiungimento di un nuovo equilibrio. A questo punto il sistema si dice "termalizzato". Come operazione preliminare è necessario quindi settare per il sistema una temperatura più alta di quella voluta, e far compiere diversi step alla simulazione finchè si osserva una stabilizzazione. Procedento per tentativi si ripete questo processo fino a che si osserva una temperatura di equilibrio il più vicina possibile alla temperatura che si desidera ottenere per il sistema.

## Studio di un sistema composto da Argon
Si procede ora con un caso specifico, quello di un sistema composto da Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ uma) in tre possibili stati (solido, liquido, gas) settando i seguenti valori:
1. **fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)**
2. **fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)**
3. **fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)**


Inoltre si considera in tutti i casi un sistema composto da $N=108$ molecole.

### Fase solida
Si riporta qui di seguito l'andamento della temperatura nei primi 2500 step della simulazione, per mostrare il comportamento sopra descritto, e la necessità di termalizzare il sistema prima di effettuare le misure delle grandezze fisiche desiderate.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creazione della figura con due subplot affiancati
fig, axes = plt.subplots(1, 2, figsize=(17,6))  
fig.suptitle("Termalizzazione del sistema", fontsize=25)

temp_evolution = np.loadtxt("./Solido/T_iniz0.8.dat", usecols=(1), unpack=True, skiprows=1)
x_axis = np.arange(1, len(temp_evolution)+1)

axes[0].set_title("T_iniz* = 0.8", fontsize=18)
axes[0].set_xlabel("Steps", fontsize=14)
axes[0].set_ylabel("T*", fontsize=14)
axes[0].hlines(0.8, 1, len(temp_evolution)+1, color='r', label="T*= 0.8")
axes[0].plot(x_axis, temp_evolution, color='b', label="T_iniz* = 0.8")
axes[0].grid(True)
axes[0].legend(loc='lower right', prop={'size': 12})


temp_evolution = np.loadtxt("./Solido/T_iniz1.545.dat", usecols=(1), unpack=True, skiprows=1)
x_axis = np.arange(1, len(temp_evolution)+1)

axes[1].set_title("T_iniz* = 1.545", fontsize=18)  
axes[1].set_xlabel("Steps", fontsize=14)
axes[1].set_ylabel("T*", fontsize=14)
axes[1].hlines(0.8, 1, len(temp_evolution)+1, color='r', label="T*= 0.8")
axes[1].plot(x_axis, temp_evolution, color='b', label="T_iniz* = 1.545")  
axes[1].grid(True)
axes[1].legend(loc='lower right', prop={'size': 12})

plt.show()
```

Come evidenziato sopra, si nota come impostando direttamente la temperatura  $T^*=0.8$ (grafico a sinistra) il sistema si termalizza ad una temperatura molto inferiore a quella desiderata, mentre impostando come valore di partenza a  $T^*=1.545$ si ottiene quato voluto.  
Dunque sono stati dapprima eseguiti 3000 steps di simulazione per portare il sistema all'equilibrio e sono state salvate le configurazioni finali di posizioni e velocità per tutte le molecole, dopodichè si è riavviata la simulazione dando in ingresso le configurazioni salvate per il sistema termalizzato.

### Fase gassosa
In questo ultimo caso studiato, nella fase di termalizzazione che viene iniziallizzata con una configurazione a reticolo si osserva un fenomeno opposto a quanto visto sopra: la temperatura di equilibrio è maggiore di quella impostata all'inizio della simulazione. La motivazione si può trovare nel fatto in un reticolo, se le particelle sono inizialmente troppo vicine (a causa della regolarità del reticolo), le forti forze repulsive le accelerano rapidamente. Questo rilascio di energia potenziale viene convertito in energia cinetica, facendo aumentare la temperatura.  

```python
import numpy as np
import matplotlib.pyplot as plt


temp_evolution = np.loadtxt("./Gas/T_iniz0.96.dat", usecols=(1), unpack=True, skiprows=1)
x_axis = np.arange(1, len(temp_evolution)+1)

plt.figure(figsize =(7,5))
plt.title("T_iniz* = 0.96", fontsize=18)  
plt.xlabel("Steps", fontsize=14)
plt.ylabel("T*", fontsize=14)
plt.hlines(1.2, 1, len(temp_evolution)+1, color='r', label="T*= 1.2")
plt.plot(x_axis, temp_evolution, color='b', label="T_iniz* = 0.96")  
plt.grid(True)
plt.legend(loc='lower right', prop={'size': 12})

plt.show()
```

Si osserva inoltre una certa instabilità del sistema nel raggiungere una temperatura di equilibrio. Servono quindi molti più steps per raggiungere una condizione di termalizzazione accettabile.
