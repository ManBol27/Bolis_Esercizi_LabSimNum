# Esercitazione 5 - Algoritmo di Metropolis

## Cenni teorici
L'algoritmo di Metropolis è di importanza fondamentale per il campionamento di valori da una certa PDF (funzioni densità di probabilità).  
Metodi come quello dell'inversa cumulativa hanno infatti un dominio di applicabilità più ristretto, dovuto per esempio all'impossibilità di invertire analiticamente la cumulativa, alla mancanza di una definizione precisa di cumulativa in spazi multidimensionali e non da ultimo la possibilità che non si riesca a calcolare la normalizzazione della PDF stessa.
Metropolis bypassa queste richieste, sfruttando invece concetti chiave dei processi stocastici quali le catene di Markov e i Random Walks.  
Partendo da un punto dello spazio delle fasi scelto in maniera opportuna (ad esempio selezionando un punto dove la PDF è massima o vicina al massimo) e utilizzando un processo iterativo, l'algoritmo dopo una fase iniziale detta di "equilibrazione", produrrà dei valori che seguono la distribuzione di probabilità desiderata.  
Più nello specifico i punti estratti seguono questa successione:
$$
d_{n+1}(x) = \int K(x|y) d_n (y) dy \; \to \; p(x)
$$
a patto che $K(x|y)$ abbia come distribuzione di probabilità strettamente positiva e invariante la $p(x)$ da cui si vuole campionare. Ciò è garantito se vale il bilancio dettagliato, ovvero se:
$$
K(x|y)p(y) = K(y|x)p(x).
$$
Per trovare una $K(x|y)$ che soddisfi quest'ultima equazione è conveniente fattorizzarla nel modo seguente:
$$
K(x|y) = T(x|y)A(x|y)
$$
dove $T(x|y)$ è da intendersi come probabilità "proposta" di transizione da $y$ a $x$, mentre $A(x|y)$ è la probabilità che la transizione proposta sia accettata.  
La scelta classica $M(RT)^2$ considera:
$$
A(x,y)=\text{min}\{1,\frac{T(x|y) \cdot p(y)}{T(y|x) \cdot p(x)}\} \quad \in [0,1] 
$$
Un'ulteriore semplificazione si ottiene considerando una $T(x|y)$ t.c. $T(x|y)=T(y|x)$, in questo modo l'algoritmo si può riassumere con i seguenti punti:
* Scelgo il punto di partenza $x_0$ in maniera opportuna come descritto sopra, una scelta meno opportuna potrebbe portare comunque a convergenza l'algoritmo anche se con un numero superiore di passi richiesti.
* Avendo selezionato una probabilità di transizione proposta simmetrica $T(x_1|x_0)$ estraggo un valore $x_1$.
* Calcolo $A(x_1|x_0)$.
* Estraggo un numero $r \in [0,1]$ da una distribuzione uniforme.
* Se $r < A(x_1|x_0)$ il nuovo valore $x_1$ è accettato altrimenti si pone $x_1 = x_0$.
* Ripetendo questi passaggi per un certo numero $M$ di volte (in questo esercizio si è scelto $M=10^4$) l'algoritmo raggiunge l'equilibrio e i numeri successivamente estratti seguiranno la distribuzione desiderata $p(x)$.

Un requisito importante per l'algoritmo di Metropolis è la scelta corretta del **passo** di transizione $\boldsymbol{S}$ , ovvero la differenza $x_{n+1}-x_n$ che dipende dalla scelta di $T(x|y)$. Un passo troppo grande porta in media a valori $p(x_{n+1})/p(x_n) \; \ll \; 1$ e di conseguenza una probabilità di accettazione della mossa molto basso, il risultato è una bassa efficenza dell'algoritmo. Per contro, passi troppo piccoli portano a $p(x_{n+1})/p(x_n) \; \sim \; 1$ e di conseguenza una probabilià di accettazione della mossa molto alta che ha come risultato un'alta correlazione tra i campionamenti e una bassa efficienza nell'esplorare tutto il dominio della PDF. Una regola empirica comunemente utilizzata è quella per cui si dovrebbe scegliere la grandezza del passo per ottenere una probabilità di accettazione $A \sim 50 \%$ dopo la fase di equilibrazione. Nel presente esercizio, procedendo per tentativi si sono trovati dei parametri efficaci per $\boldsymbol{S}$ nell'ottenere quanto appena descritto.

## Es 5.01 - Atomo di idrogeno con transizione uniforme
Nel presente esercizio si è implementato un algoritmo di Metropolis per lo studio della densità di probabilità associata alla funzione d'onda dell'atomo di idrogeno nei due stati GS e 2P (modulo quadro della funzione d'onda).Si parte dunque dal risultato analitico:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
per il ground state, mentre per lo stato eccitato $2P$:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta).
$$
Si è scelta inizialmente come probabilità di transizione proposta una **$T(x|y)$ uniforme** nelle tre coordinate cartesiane, ovvero $T(\vec{x}_{n+1}|\vec{x}_n)$ è tale che 
$$ \vec{x}_{n+1} = (x^1_{n+1}, \; x^2_{n+1}, x^3_{n+1}) = (x^1_n, x^2_n , x^3_n) + S \; (U^1, U^2, U^3)
$$ , dove $U^i = \text{unif}[-1,1]$. 

<i>Da notare che in questo modo il passo non ha modulo costante, dunque soddisfa la condizione di non essere troppo elevato ma al contempo statisticamente assumerà valori anche prossimi a zero con la possibilità di campionare a volte valori correlati tra di loro. Si potrebbe dunque lavorare in coordinate sferiche e fissare il modulo $r$ del passo, in maniera tale da variare uniformemente soltanto la direzione (vedi esercitazione precedente sui random walks). Nonostante ciò, il codice ha prodotto dei buoni risultati e si è pertanto deciso di mantenere questa scelta che semplifica l'implementazione. </i>

Si mostrano ora i risultati ottenuti per i valori di **raggio medio** estraendo le posizioni dalla PDF del ground state e poi da quella dello stato 2P. Si ricordi che analiticamente ci si aspetta che:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
I risultati mostrati sono espressi in unità del raggio di Bohr $(a_0 =1)$. 

In entrambi i casi si è scelto di svolgere $M=10^4$ passi di equilibrazione, per poi campionare $10^6$ valori di posizione in 3D e calcolare la solita media cumulativa a blocchi. 

Per entrambi i casi si è scelto di impostare come valore di posizione iniziale $x_0 = (0,0,0)$, scelta ponderata per il caso GS mentre non ottimale per il caso 2P (che ha un massimo della PDF in $(0,0,2)$). Si può vedere facendo alcune simulazioni che effettuando sufficienti passi per equilibrare il sistema, la scelta del valore iniziale (a patto che non sia eccessivamente lontano dai punti in cui la PDF è massima), non è determinante per il risultato finale.

### Risultati per $\langle r \rangle_{GS}$


```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(23,9))  
fig.suptitle("Confronto del valore $<r>_{GS}$ prima e dopo equilibrazione", fontsize=25)

print("I valori medi osservati sono:", "\n")

ave, err = np.loadtxt("Uniforme/GS_Equilibration_1.210000.txt", usecols=(0,1), unpack=True, skiprows=1)
x_axis = np.arange(1, len(ave)+1)


axes[0].set_title("$<r>_{GS}$ - Equilibrazione", fontsize=21)
axes[0].set_xlabel("Blocchi", fontsize=19)
axes[0].set_ylabel("<r>", fontsize=19)
axes[0].errorbar(x_axis, ave, err, color='b', label="<r>")
axes[0].hlines(1.5, 0, len(ave)+1, color='r', label="Valore atteso")
axes[0].grid(True)
axes[0].legend(loc='lower right', prop={'size': 12})


ave, err = np.loadtxt("Uniforme/GS_1.210000.txt", usecols=(0,1), unpack=True, skiprows=1)
x_axis = np.arange(1, len(ave)+1)

axes[1].set_title("$<r>_{GS}$ - Dopo equilibrazione", fontsize=21)
axes[1].set_xlabel("Blocchi", fontsize=19)
axes[1].set_ylabel("<r>", fontsize=19)
axes[1].errorbar(x_axis, ave, err, color='b', label="<r>")
axes[1].hlines(1.5, 0, len(ave)+1, color='r', label="Valore attesp")
axes[1].grid(True)
axes[1].legend(loc='lower right', prop={'size': 12})

print("Raggio medio: ", np.round(ave[-1],4)," +- ", np.round(err[-1],4), "\n" )
print("In accordo con il valore atteso entro", np.round(np.abs(ave[-1]-1.5)/err[-1], 2), "sigma", "\n")
```

Il confronto tra i due grafici evidenzia come la fase di equilibrazione sia fondamentale per osservare dei risultati sensati, in accordo con il valore atteso. 
Si è utilizzato un passo $S = 1.21$ che ha restituito i seguenti acceptance rates:
* 50.19 % in media nella fase di equilibrazione di $10^4$ steps
* 50.41 % in media dopo l'equilibrazione, nei successivi $10^6$ steps.

## Es 5.02 - Atomo di idrogeno con transizione gaussiana
In questa seconda parte viene eseguito il medesimo studio effettuato sopra, utilizzando però una probabilità di transizione proposta $T(y|x)$ gaussiana. In questo caso se il passo viene accettato (mediante la stessa $A(x|y)$ descritta sopra) la successione degli $x_n$ ha segue l'espressione:
$$ 
\vec{x}_{n+1} = (x^1_{n+1}, \; x^2_{n+1}, x^3_{n+1}) = (x^1_n, x^2_n , x^3_n) + S \; (N^1, N^2, N^3)
$$ , dove $N^i$ è la distribuzione normale scelta con media nulla e varianza pari a $\delta^2$.

In questo caso quindi, quello che sopra è stato definito "passo" ora coincide con la radice quadrata della varianza, ovvero:
$$
S \equiv \delta
$$

Stampo ora $2 \times 10^5$ punti estratti dalla PDF dello stato 2P per osservare come si distibuiscono nello spazio 3D:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import imageio.v2 as imageio
import os

# === Parametri ===
file_input = "Gauss/Punti_estratti_2P.txt"
gif_output = "rotazione.gif"
n_frames = 10
fps = 0.5


# === Carica dati (salta intestazione) ===
data = np.genfromtxt(file_input, skip_header=1)

# === Crea lista in ci salvare le immagini ===
filenames = []

# Creo un vettore di raggi, calcolato come norma (aka modulo) per ogni riga (axis=1 significa che opera colonna per colonna)
radii = np.linalg.norm(data, axis=1)

for i, angle in enumerate(np.linspace(0, 360, n_frames)):
    fig = plt.figure(figsize=(6, 5))
    ax = fig.add_subplot(111, projection='3d')

    #c=data... per colorare i dati in base al valore di r
    scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=radii[:], cmap='viridis', s=30)
    ax.view_init(elev=angle, azim=45)


    # Aggiungi la barra dei colori per i raggi
    plt.colorbar(scatter, ax=ax)

    # Aggiungi le frecce per asse z
    ax.quiver(0, 0, 0, 0, 0, 1, length=23, color='r', arrow_length_ratio=0.1, linewidth=2) 
    
    filename = f"frame_{i:03d}.png"
    plt.savefig(filename)
    filenames.append(filename)
    plt.close()


# === Crea GIF ===
with imageio.get_writer(gif_output, mode='I', duration=1500) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)



print(f"GIF salvata come {gif_output}")

```
