# Esercitazione 6 - Modello di Ising

## Cenni teorici al modello

### Ising 1D

Il modello di Ising è un modello utilizzato per descrivere le proprietà magnetiche dei materiali, l'idea su cui si basa è quella di considerare un blocco di materiale composto da un reticolo (di qualsiasi geometria) dove in ogni punto di quest'ulitmo è localizzato uno spin (che nell'ipotesi 1D è uno scalare che può assumere i valori $s_i = \pm 1$
Il modello ha soluzione analitica in 1D e 2D ma non in 3D. Nella seguente esercitazione si sono simulate delle misure di grandezza fisiche per un cristallo composto da $N = 50$ spin in 1D a certe temperature fissate $T = \frac{1}{k_B \beta}$, confrontadole poi con i risultati analitici. Essendo gli spin localizzati nei siti del reticolo, non è presente una dinamica del sistema, e l'Hamiltoniana assume la seguente forma:

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove il primo termine rappresenta l'interazione tra spin primi vicini scalata su $J$ (parametro di interazione,è positivo se il materiale è ferromagnetico e quindi favorisce l'allineamento tra gli spin abbassando l'energia, negativo se il materiale è antiferromagnetico), mentre il secondo termine contiene $h$, ovvero il campo magnetico esterno che sarà preso uguale a zero, tranne che per il calcolo della magnetizzazione che altrimenti risulterebbe nulla.  
Inoltre sono state imposte delle condizioni al contorno tali da rappresentare un sistema periodico (aka $s_{N+1} = s_1$), per far sì che ogni spin avesse lo stesso numero di primi vicini. 

Dalla meccanica statistica viene che la densità di probabilità che il sistema si trovi in un certo **stato $\boldsymbol{\mu}$** con una certa energia $H_{\mu}$ è data da:
$$
p(\mu) = \frac{\exp \left( -\beta H\right)}{Z}
$$

con $Z$ funzione di partizione. Da questa espressione si può ricavare il valore medio di una certa grandezza fisica $G$ con la seguente formula:
$$
\langle G \rangle = \sum_{\mu} G(\mu)p(\mu)
$$

E' a questo punto che entrano in gioco gli algoritmi di Metropolis e di Gibbs per campionare da una densità di probabilità qualunque anche senza conoscerne la normalizzazione.

*I risultati riportati di seguito sono espressi in unità di $k_B$ e di  $\mu_b$.*

### Ising 1D - Metropolis
L'algoritmo di Metropolis è già stato discusso in dettaglio nell'esercitazione precedente, segue pertanto l'idea di come si applica al caso specifico del modello di Ising in 1D. 
Si parte da una configurazione per $N$ spin, ciascuno dei quali assume un valore inizialmente casuale ($\pm1$), che rappresenta l'ipotesi di $T_{iniz} = \infty$. Il primo passo consiste nel generare un nuovo stato $\nu$ a partire dal presente stato $\mu$, invertendo il valore di un singolo spin (scelto casualmente) secondo una probabilità di transizione proposta $T(\nu | \mu)$. La mossa viene poi accettata o meno con probabilità data da:
$$
A(\mu | \nu) = \text{min} \left [ 1, \frac{T(\mu | \nu) p(\nu)}{T(\nu | \mu) p(\mu)} \right ]
$$
che nel caso in cui si ha simmetria della probabilità di transizione, ovvero $T(\mu | \nu) = T(\nu | \mu)$ si semplifica a:
$$
A(\mu | \nu) = \text{min} \left [ 1, \frac{p(\nu)}{p(\mu)} \right ] = \text{min} \left [ 1, e^{-\beta (E_{\nu}-E_{\mu})} \right ].
$$

Nel caso di *campo magnetico nullo* la differenza di energia si calcola come:
$$
E_{\nu}-E_{\mu} = 2 J s_i^{\mu} \sum_{\text{k primi vicini}} s_k^{\mu}
$$
dove $s_i^{\mu}$ è lo spin che si sta considerando di flippare.

**Si noti la dipendenza fondamentale dalla temperatura $\boldsymbol{T}$.**

Uno step consiste quindi nel considerare randomicamente N spin (non per forza prendendo tutti quelli del sistema,proprio perchè presi randomicamente, con la possibilità quindi di ripetere più volte l'operazione sullo stesso spin), ciascuno dei quali viene flippato o meno in base a quanto descritto sopra. Terminato uno step, si calcolano le grandezze fisiche di interesse sulla configurazione attuale degli $N$ spin.

Si procede poi con il solito block averaging, considerando $40$ blocchi da $10^5$ simulazioni ciascuno. Volendo costruire un andamento del tipo $G(T)$ delle grandezze al variare della temperatura, l'operazione consiste nel fissare un valore di $T$ alla volta e prendere come risultato per una certa grandezza fisica, il valore medio cumulativo dopo l'ultimo blocco considerato.

### Ising 1D - Gibbs 
L'algoritmo di Gibbs (anche detto "heat bath algorithm") è un metodo efficace per campionare da distribuzioni di probabilità multivariate. L'idea di base è campionare una variabile alla volta (nel nostro modello significa considerare uno spin alla volta) utilizzando la probabilità condizionata sui valori correnti delle altre. 

Si può vedere l'algoritmo di Gibbs come un caso particolare del Metropolis, dove le transizioni proposte sono sempre accettate ($A(\mu | \nu) = 1$). Più in dettaglio, nel caso di Ising 1D si comincia col fabbricare uno stato a $N = 50$ spin orientati casualmente, dopodichè il primo passo è costruire una nuova configurazione mediante i seguenti passi:
* Si considera uno degli spin scelto a random
* Si calcola la probabilità che questo si trovi in uno stato spin up, condizionata dallo stato degli altri spin (nella pratica con l'interazione considerata, ci interessano solo i primi vicini):
$$
p(s_k' = \pm 1 \mid \{s_j : j \ne k\}) = 
\frac{1}{1 + \exp\left( -\beta \Delta E \Big|_{s_k = \pm1 \to s_k = \mp1} \right)} = 
\frac{1}{1 + \exp\left( \mp 2 \beta J \sum_{\langle i \in \text{n.n. to } k \rangle} s_i \right)}
$$
* Si estrae un numero casuale $r$ in $[0,1]$, se $r < p(s_k' = \pm 1 \mid \{s_j : j \ne k\})$ allora si pone lo spin $s_k = 1$, altrimenti si pone $s_k = -1$
* Si ripetono queste operazioni per un numero di volte pari a N, ottenendo quindi una nuova configurazione.

Si procede infine con il solito block averaging utilizzando $40$ blocchi con $10^5$ steps (configurazioni) ciascuno.


***Note sull'equilibrazione del sistema*:** 

Dalle varie simulazioni effettuate si è notato come al sistema non occorrano molti steps per raggiungere una condizione di stabilità, pertanto si è decisio di non effettuare una equilibrazione preliminare. Inoltra si è osservato come dopo circa $25$ blocchi la media cumulativa di una certa grandezza si assesta, questo giustifica la scelta dei 40 blocchi descritta sopra. Si riporta a titolo di esempio, il calcolo della magnetizazzione a $T=2$ con $40$ blocchi da $10^5$ simulazioni ciascuno dove si può notare il comportamento appena descritto:

```python
import numpy as np
import matplotlib.pyplot as plt

ave, err = np.loadtxt("magnetization_example.dat", unpack= True, usecols=(2,3), skiprows=1)

N_blocks = len(ave) # numero blocchi
x = np.arange(1, N_blocks+1)

plt.figure(figsize=(7,5))

plt.title(r"Magnetizzazione a T=2", fontsize=15)

plt.errorbar(x, ave, yerr=err, label = "Magnetizzazione", color = 'orange') 
plt.xlabel(r"Blocchi", fontsize=12)
plt.ylabel(r"Magnetizzazione", fontsize=12)

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.legend(loc='lower right', prop={'size': 12})

plt.grid(True)

plt.show()
```

### Grandezze fisiche considerate: 
Nella presente esercitazione si sono stimate per un sistema di $N=50$ spin, le seguenti grandezze fisiche al variare della temperatura $T$ nell'intervallo $[0.5, 2.0]$:


1. Energia interna $U(N,T) = \langle H \rangle_{h=0}$
2. Calore specifico $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. Suscettibilità magnetica $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. Magnetizzazione $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

Viene considerato il caso $J=1$, inoltre vengono dati i risultati delle quantità appena elencate riscalate sul numero di spin.

Di seguito i risultati ottenuti:

```python
import numpy as np
import matplotlib.pyplot as plt

m_energy, m_energy_err = np.loadtxt("Metropolis_Data/tot_energy(T).dat", unpack= True, usecols=(1,2), skiprows=1)
m_sh, m_sh_err = np.loadtxt("Metropolis_Data/Specific_Heat(T).dat", unpack= True, usecols=(1,2), skiprows=1)
m_suscept, m_suscept_err = np.loadtxt("Metropolis_Data/Susceptibility(T).dat", unpack= True, usecols=(1,2), skiprows=1)
m_magne, m_magne_err = np.loadtxt("Metropolis_Data/magnetization(T)_h002.dat", unpack= True, usecols=(1,2), skiprows=1)

g_energy, g_energy_err = np.loadtxt("Gibbs_Data/tot_energy(T).dat", unpack= True, usecols=(1,2), skiprows=1)
g_sh, g_sh_err = np.loadtxt("Gibbs_Data/Specific_Heat(T).dat", unpack= True, usecols=(1,2), skiprows=1)
g_suscept, g_suscept_err = np.loadtxt("Gibbs_Data/Susceptibility(T).dat", unpack= True, usecols=(1,2), skiprows=1)
g_magne, g_magne_err = np.loadtxt("Gibbs_Data/magnetization(T)_h002.dat", unpack= True, usecols=(1,2), skiprows=1)


# Parametri analitici
points = 100
T = np.linspace(0.5, 2.0, num=points)
beta = 1/T
J = 1.0
Ns = 50

# Energia interna analitica
th = np.tanh(J/T)
thN = th**Ns
ch = 1/th
e = -J * (th + ch * thN) / (1 + thN)

# Calore specifico analitico
heat = ((beta*J)**2) * (
    ((1 + thN + (Ns - 1)*(th**2) + (Ns - 1)*(ch**2)*thN) / (1 + thN)) -
    Ns*((th + ch*thN)/(1 + thN))**2
)

# Magnetizzazione analitica (con campo esterno h = 0.02)
h = 0.02
b = 1/T
cosh_bh = np.cosh(b*h)
sinh_2bJ = np.sinh(2*b*J)
sqrt_term = np.sqrt(np.exp(2*b*J)*cosh_bh**2 - 2*sinh_2bJ)

l1 = np.exp(b*J)*cosh_bh + sqrt_term
l2 = np.exp(b*J)*cosh_bh - sqrt_term
Z = l1**Ns + l2**Ns

term1 = l1**(Ns-1) * (1 + np.exp(b*J)*cosh_bh / sqrt_term)
term2 = l2**(Ns-1) * (1 - np.exp(b*J)*cosh_bh / sqrt_term)

M = (np.exp(b*J)*np.sinh(b*h)*(term1 + term2)) / Z
plt.figure(figsize=(17,13))

#Suscettibilità analitica
X = beta * np.exp(2*beta*J) * (1 - thN) / (1 + thN)

#---GRAFICI---

plt.suptitle("Confronto tra algoritmi Metropolis e Gibbs nel modello di Ising 1D", fontsize=22)
plt.figure(figsize=(12, 13)) 


# Grafico Energia interna

plt.subplot(2,2,1)

plt.title(r"Energia interna per spin", fontsize=18)
plt.xlim(0.4, 2.1)

#Risultati numerici
plt.errorbar(x, m_energy, yerr=m_energy_err, fmt='D', label = "Metropolis", color = 'orange')
plt.errorbar(x, g_energy, yerr=g_energy_err, fmt='o', label = "Gibbs", color = 'green')
#Risultati analitici
plt.plot(T, e, label = "Analitico")

plt.xlabel('T', fontsize = 14)
plt.ylabel('U/N', fontsize = 14)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)


plt.legend(loc='lower right', prop={'size': 12})
plt.grid(True)



# Calore specifico
plt.subplot(2,2,2)
plt.title(r"Calore specifico per spin", fontsize=18)
plt.xlim(0.4, 2.1)

plt.errorbar(x, m_sh, yerr=m_sh_err, fmt='D', label="Metropolis", color='orange')
plt.errorbar(x, g_sh, yerr=g_sh_err, fmt='o', label="Gibbs", color='green')
plt.plot(T, heat, label="Analitico", color='blue')

plt.xlabel('T', fontsize=14)
plt.ylabel(r"$C_V/N$", fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc='upper right', prop={'size': 12})
plt.grid(True)

# Suscettibilità magnetica 
plt.subplot(2,2,3)
plt.title(r"Suscettibilità magnetica per spin", fontsize=18)
plt.xlim(0.4, 2.1)
plt.ylim(-10, 100)

plt.errorbar(x, m_suscept, yerr=m_suscept_err, fmt='D', label="Metropolis", color='orange')
plt.errorbar(x, g_suscept, yerr=g_suscept_err, fmt='o', label="Gibbs", color='green')
plt.plot(T, X, label="Analitico", color='blue')

plt.xlabel('T', fontsize=14)
plt.ylabel(r"$\chi /N$", fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc='upper right', prop={'size': 12})
plt.grid(True)

# --- Magnetizzazione ---
plt.subplot(2,2,4)
plt.title(r"Magnetizzazione per spin (h = 0.02)", fontsize=18)
plt.xlim(0.4, 2.1)

plt.errorbar(x, m_magne, yerr=m_magne_err, fmt='D', label="Metropolis", color='orange')
plt.errorbar(x, g_magne, yerr=g_magne_err, fmt='o', label="Gibbs", color='green')
plt.plot(T, M, label="Analitico", color='blue')

plt.xlabel('T', fontsize=14)
plt.ylabel(r"$M/N$", fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc='upper right', prop={'size': 12})
plt.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])  # lascia spazio per il titolo in alto
plt.show()
```

Come si osserva dalle figure, entrambi i metodi hanno fornito risultati perfettamente compatibili con le soluzioni analitiche. In particolare l'energia interna aumenta con la temperatura, in quanto la configurazione degli spin si allontana da quella di spin allineati tra di loro. Per lo stesso motivo la magnetizzazione, che in qualche modo esprime il "grado di allineamento" degli spin tende a zero.
