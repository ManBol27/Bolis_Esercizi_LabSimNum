## Ex 01.1 - Generatore di numeri casuali con distribuzione uniforme

### Media

Si vuole dapprima testare il generatore di numeri casuali fornito, a questo scopo si calcola una media integrale del tipo:
$$
\langle r \rangle = \int_0^1 r dr = 1/2
$$
Siccome però i numeri generati non sono propriamente casuali, poichè generati da un algoritmo per sua natura deterministico, ma **"pseudocasuali"** (dunque correlati tra di loro), per una valutazione corretta della bontà del generatore si calcola una media su blocchi con un numero pari a L di numeri generati. Questa prcedura permette di ridurre la correlazione tra i numeri con cui si va a calcolare la media statistica 
Si è proceduto dunque con il calcolo della media con relativo errore su N=100 blocchi, ciascuno contenente un numero L=10000 estrazioni, per un totale di $10^6$ numeri generati.  
Il codice è stato scritto in linguaggio C++ e ha prodotto i seguenti risultati:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Per importare i dati uso la funzione "loadtxt" di Numpy
ave, err = np.loadtxt("text1.txt", unpack=True, usecols=(0,1), skiprows=1)

#unpack=True per poter salvare le diverse colonne in diverse variabili;  usecols=(0,1) per indicare quali colonne utilizzare; 
#skiprows=1 per saltare la prima riga che contiene le intestazioni delle colonne
```

```python
N=100 #numero di blocchi
x_axis = np.arange (1, N+1)
plt.figure(figsize=(8, 6))
plt.title("Media cumulativa per l'estrazione di numeri casuali per distribuzione uniforme", fontsize=17, weight="bold")
plt.errorbar(x_axis,ave-0.5,yerr=err, label="Simulato")
plt.xlabel(r"# Blocchi", fontsize=15)
plt.ylabel(r"<r>-1/2", fontsize=15)
plt.axhline( y=0, xmin=0, xmax=100, color='red', label="Val. atteso")
plt.legend()
plt.grid(True)
plt.show()

print("La media cumulativa dopo 100 blocchi è: ", ave[-1], "con un errore di: ", err[-1], ", ovvero entro ", round((ave[-1]-0.5)/err[-1],2), "sigma.")
```



La bontà del generatore di numeri casuali utilizzato è dunque confermata.

### Deviazione standard

Si procede ora sulla falsa riga di quanto elaborato sopra considerando blocchi di 10000 estrazioni, per calcolare la deviazione standard dalla media con relativo errore. Di nuovo considerando la media intesa come media integrale ci si aspetta un valore pari a $1/12$ per la dev. standard:
$$
\sigma^2 = \int_0^1 (r-1/2)^2 dr = \frac{1}{12}
$$
e utilizzando gli stessi numeri generati al punto precedente si ottiene:

```python
#Carico i dati in due nuovi vettori
ave_ds, err_ds = np.loadtxt("text2.txt", unpack=True, usecols=(0,1), skiprows=1)

N=100 #numero di blocchi
x_axis = np.arange (1, N+1)
plt.figure(figsize=(9, 7))
plt.title("Media cumulativa della dev. standard per l'estrazione di numeri casuali per distribuzione uniforme",  fontsize=17, weight="bold")
plt.errorbar(x_axis,ave_ds-1/12,yerr=err_ds, label="Simulato")
plt.xlabel(r"# Blocchi", fontsize=15)
plt.ylabel(r"$\langle \left(r-0.5 \right)^{2} \rangle$ - $\frac{1}{12}$", fontsize=15)
plt.axhline( y=0, xmin=0, xmax=100, color='red', label="Val. atteso")
plt.legend()
plt.grid(True)
plt.show()

print("La dev. standard cumulativa dopo 100 blocchi è: ", ave_ds[-1], 
      "con un errore di: ", err_ds[-1], ", ovvero entro ", round((ave_ds[-1]-1/12)/err_ds[-1],2), "sigma.")
```

### Test del $\chi^{2}$

Si vuole ora eseguire un test del $\chi^{2}$ per l'ipotesi che i numeri generati **non** siano effettivamente appartenenti ad una distribuzione uniforme.
Considerando che i numeri generati sono compresi nell'intervallo $[0,1]$ è possibile suddividere quest'ultimo in un numero pari a $n_{subint}=100$ di sottointervalli. In ciascuno degli $N=100$ blocchi in cui si sono raggruppati le estrazioni effettuate in precedenza ci si aspetta che gli $L=10000$ numeri generati si distribuiscano in maniera uniforme nei $n_{subint}=100$ sottointervalli, ovvero che ciascuno di questi ultimi contenga in media $L/n_{subint}=100$ estrazioni. 
<br>
Per ogni blocco si può dunque calcolare il $\chi^{2}$ usando la formula di Pearson:
$$
\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}
$$
dove $O_i$ è il numero di elementi effettivamente contenuti nel i-esimo sottointervallo, mentre il numero aspettato è $E_i=100$ per quanto scritto sopra.
<br>
Siccome la probabilità che un lancio cada in un certo sottointervallo è data da $p=\frac{1}{n_{subint}}$, quella che si sta considerando è una distribuzione multinomiale (analoga alla binomiale ma con più di 2 possibili esiti di un'estrazione), che al limite tende alla distribuzione di Poisson per la quale vale che varianza$\sim$media. Per questo motivo nella formula di Pearson scritta sopra si può considerare che il numeratore (i.e. varianza$\sim$media) e il denominatore (media) siano circa uguali, dunque il loro rapporto è 1 e il valore atteso per il $\chi^{2}$ è $100$.
Un valore effettivo del $\chi^{2}$ che supera di molto il valore atteso indicherebbe che l'ipotesi che le sole fluttuazioni casuali spieghino i dati diventa improbabile e potremmo concludere che il generatore di numeri casuali non estragga veramente numeri da una distribuzione uniforme. Viceversa un valore basso del $\chi^{2}$  indicherebbe che quest'ultima ipotesi è scartabile.

```python
## Carico i dati in due nuovi vettori
chi2 = np.loadtxt("text3.txt", unpack=True, usecols=(0), skiprows=0)

media_chi=np.mean(chi2)
plt.figure(figsize=(9,8))
plt.hist(chi2, bins=15, density=True)   #density=True serve per far si che l'area dell'istogramma sia normalizzata a 1
plt.axvline(media_chi, label=rf'Media $\chi^2$ = {round(media_chi,2)}', color="red", ls="-.", lw=2)
plt.axvline(100, label=rf'Media aspettata $\chi^2$ = 100', color="green", ls="--", lw=2)
plt.legend(loc='upper right', prop={'size': 12})
plt.title(r'Distribuzione $\chi^2$ per N Blocchi', fontsize=18)
plt.xlabel(r'$\chi^2$', fontsize=15)
plt.ylabel("Estrazioni", fontsize=15)
plt.grid(True)
plt.show()

print(rf'Il χ² mediato su N blocchi è: ', round(media_chi,2))
```

Il test effettuato verifica che in effetti è invalidata l'ipotesi per la quale i numeri non sarebbero effettivamente estratti da una distribuzione uniforme.

## Ex 01.2 - Generatore di numeri casuali con distribuzioni alternative alla uniforme

Si può realizzare un generatore di numeri casuali estratti da distribuzioni alternative a quella uniforme utilizzando il teorema dell'inversa cumulativa, il cui statement è il seguente:

Se $X$ è una variabile casuale con funzione cumulativa:
$$
F(X) = \int_{-\infty}^{X} p(X') dX'
$$
allora la variabile casuale $X=F^{-1}(U)$ segue la distribuzione $p(X)$ dove $U$ è una variabile casuale uniformemente estratta dall'intervallo $[0,1)$, e $F^{-1}$ è l'inversa della funzione cumulativa. 
<br>
Si sono considerate nello specifico:
* Distribuzione esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$, con inversa della funzione cumulativa
  $$x = -\frac{1}{\lambda} \cdot \ln(1-U) \quad$$
* Distribuzione lorentziana $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$ , con inversa della funzione cumulativa
  $$x = \Gamma \cdot \tan\left[\pi\left( U-\frac{1}{2} \right) \right] + \mu \quad $$



Vogliamo con queste distribuzioni verificare il **TEOREMA DEL LIMITE CENTRALE** il cui statement è:  
Sia $ X_1, X_2, ..., X_n $ una sequenza di **n** variabili casuali indipendenti e identicamente distribuite (i.i.d.) con:

- media \( $\mu$ \),
- varianza finita \( $\sigma^2$ \).

Allora, la somma normalizzata converge alla distribuzione normale per \( $N \to +\infty $\):

$$
S_N = \frac{1}{N}\sum_{i=1}^N x_i \to N(0,1)
$$

Si considerano dunque $10^4$ estrazioni della varibile statistica $S_N$ costruita con $N=1, 2, 10, 100$ numeri casuali $X_i$ e si plotta un istogramma per leggerne la distribuzione. Se il teorema del limite centrale è verificato, il plot dell'istogramma per $S_N$ dovrebbe tendere ad una distribuzione gaussiana all'aumentare di $N$.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import cauchy

#Creo l'istogramma per distribuzione uniforme
S_N1, S_N2, S_N10, S_N100 = np.loadtxt("S_N_unif.txt", unpack=True, usecols=(0,1,2,3), skiprows=(1))
plt.figure(figsize=(9,8))
plt.hist(S_N1, bins=40, density=True, color="blue", alpha=0.7, label='N=1')
plt.hist(S_N2, bins=40, density=True, color="green", alpha=0.7, label='N=2')
plt.hist(S_N10, bins=40, density=True, color="yellow", alpha=0.7, label='N=10')
plt.hist(S_N100, bins=40, density=True, color="red", alpha=0.7, label='N=100')

#Usando l'ultimo istogramma con N=100 costruisco una gaussiana con la stessa media e dev. standard dei dati dell'istogramma
media=np.mean(S_N100)
devst=np.std(S_N100)
x = np.linspace(min(S_N100), max(S_N100), 100)
y = norm.pdf(x, media, devst) #norm.pdf è una funzione che genera la densità di probabilità (PDF) della gaussiana

plt.plot(x, y, color="black", label=f'Fit Gaussiano\nμ={media:.2f}, σ={devst:.2f}')  #.2f vuol dire con due cifre decimali

plt.legend(loc='upper right', prop={'size': 12})
plt.title(r'Distribuzione di $S_N$ da probabilità uniforme', fontsize=18)
plt.xlabel('$S_N$', fontsize=15)
plt.ylabel('Estrazioni', fontsize=15)
plt.grid(True)
plt.show()

#Creo l'istogramma per distribuzione esponenziale
S_N1, S_N2, S_N10, S_N100 = np.loadtxt("S_N_expo.txt", unpack=True, usecols=(0,1,2,3), skiprows=(1))
plt.figure(figsize=(9,8))
plt.hist(S_N1, bins=40, density=True, color="blue", alpha=0.7, label='N=1')
plt.hist(S_N2, bins=40, density=True, color="green", alpha=0.7, label='N=2')
plt.hist(S_N10, bins=40, density=True, color="yellow", alpha=0.7, label='N=10')
plt.hist(S_N100, bins=40, density=True, color="red", alpha=0.7, label='N=100')

#Usando l'ultimo istogramma con N=100 costruisco una gaussiana con la stessa media e dev. standard dei dati dell'istogramma
media=np.mean(S_N100)
devst=np.std(S_N100)
x = np.linspace(min(S_N100), max(S_N100), 100)
y = norm.pdf(x, media, devst) #norm.pdf è una funzione che genera la densità di probabilità (PDF) della gaussiana

plt.plot(x, y, color="black", label=f'Fit Gaussiano\nμ={media:.2f}, σ={devst:.2f}')  #.2f vuol dire con due cifre decimali

plt.legend(loc='upper right', prop={'size': 12})
plt.title(r'Distribuzione di $S_N$ da probabilità esponenziale', fontsize=18)
plt.xlabel('$S_N$', fontsize=15)
plt.ylabel('Estrazioni', fontsize=15)
plt.grid(True)
plt.show()

#Creo l'istogramma per distribuzione lorentziana
S_N1, S_N2, S_N10, S_N100 = np.loadtxt("S_N_lorentz.txt", unpack=True, usecols=(0,1,2,3), skiprows=(1))
plt.figure(figsize=(9,8))
plt.hist(S_N1, bins=np.linspace(-15,15,60), density=True, color="blue", alpha=0.7, label='N=1')
plt.hist(S_N2, bins=np.linspace(-15,15,60), density=True, color="green", alpha=0.7, label='N=2')
plt.hist(S_N10, bins=np.linspace(-15,15,40), density=True, color="yellow", alpha=0.7, label='N=10')
plt.hist(S_N100, bins=np.linspace(-15,15,40), density=True, color="red", alpha=0.7, label='N=100')
plt.xlim(-15,15)
#Usando l'ultimo istogramma con N=100 costruisco una gaussiana con la stessa media e dev. standard dei dati dell'istogramma
media, devst=cauchy.fit(S_N100)
x = np.linspace(-15, 15, 100)
y = cauchy.pdf(x, media, devst) #cauchy.pdf è una funzione che genera la densità di probabilità (PDF) della cauchy

plt.plot(x, y, color="black", label=f'Fit Cauchy\nμ={media:.2f}, σ={devst:.2f}')  #.2f vuol dire con due cifre decimali

plt.legend(loc='upper right', prop={'size': 12})
plt.title(r'Distribuzione di $S_N$ da probabilità lorentziana', fontsize=18)
plt.xlabel('$S_N$', fontsize=15)
plt.ylabel('Estrazioni', fontsize=15)
plt.grid(True)
plt.show()
                                  
```

Dai grafici ottenuti si evidenzia che per la distribuzione uniforme e per quella esponenziale la distribuzione della variabile aleatoria $S_N$ tende in effetti ad una distribuzione normale.  
Nel caso di numeri estratti da una distribuzione lorentziana invece si ottiene un risultato diverso: essendo che la **lorentziana non ha varianza finita**, le ipotesi del teorema del limite centrale classico non sono soddisfatte. Dal punto di vista teorico la distribuzione lorentziana è già una distribuzione limite, difatto non si vedono variazioni del grafico all'aumentare di N (per questo motivo per l'ultimo grafico è stato fatto un fit con Cauchy).

## Ex 01.3 - Simulazione esperimento di Buffon per stima del $\pi$

L'esperimento di Buffon consiste nel seguente problema: un ago di lunghezza $L$ è lanciato in maniera casuale su un piano orizzontale avente un pattern con delle sottili linee verticali poste ad una distanza $D$ una dall'altra.  Sia $L$ la lunghezza dell'ago (con $D>L$ ma non $D\gg L$). La probabilità che l'ago intersechi una di queste linee è $P=\frac{2L}{\pi D}$, dunque si può invertire quest'ultima formula per ricavare il valore di $\pi$ conoscendo la probabilità $P=\lim_{N_{lanci} \to \infty} \frac{N_{lanci}}{N_{inter}}$. Una stima del $\pi$ è data dunque da:
$$
\pi = \frac{2L}{Pd} = \lim_{N_{lanci} \to \infty}\frac{2LN_{lanci}}{N_{inter}d}
$$

Nello spirito del primo esercizio, si costruisce quindi la media cumulativa con relativo errore considerando un numero pari a $M$ di lanci raggruppati in un numero $N$ di blocchi.

In dettaglio la procedura seguita per determinare se un lancio interseca o meno una linea è il seguente:  

L'idea è estrarre due punti per definire implicitamente un angolo di orientazione dell'ago senza usare esplicitamente il $\pi$ (che devo invece ricavare). Infatti si potrebbe semplicemente pensare di estrarre una coordinata e un angolo da una distribuzione uniforme di valori compresi tra $[0,2\pi)$, questo approccio però avrebbe poco senso siccome si utilizzerebbe il valore di $\pi$ che si considera in questo problema ignoto.  
Devo dunque procedere per i seguenti punti:
- Estrarre solo la coordinata $x_a$ del primo punto (centro della circonferenza in figura) da una distribuzione uniforme tra $L$ e $L+D$, siccome posso supporre il pattern infinito in altezza (assumo $y_a=0$).
- Estraggo le coordinate del secondo punto da distribuzione uniforme: estraggo $x_b$ dall'intervallo $[x_a-L , x_a+L]$ e a coordinata $y_b$ dall'intervallo $[-L , L]$
- I numeri estratti in questa maniera non assegnano ad ogni direzione la stessa probabilità: ci sono delle direzioni privilegiate (come si può osservare intuitivamente dalla figura ci sono più punti del quadrato orientati a $-45°$ rispetto a quelli orientati orizzontalmente a $0°$). Per ristabilire l'equiprobabilità nell'estrazione della direzione dell'ago, si possono semplicemente scartare tutte le coordinate estratte che hanno una distanza dal centro superiore a L (in questo modo sto considerando i punti nella circonferenza inscritta nel quadrato, che non privilegia certi angoli).
- Una volta definita una direzione ridefinisco un nuovo punto $(x^*_b , y^*_b)$ che sia distante $L$ dal punto $P_a = (x_a , y_a)$ lungo quella direzione.
- Ora se $x^*_b > L+D$ oppure $x^*_b < L$ allora l'ago interseca le linee del pattern.
