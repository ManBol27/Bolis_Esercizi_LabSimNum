## Ex 7.01 - Tail corrections
Nell'esercizio 4 si è simulato il comportamento di un sistema molecolare a energia fissata. In prima istanza si è scelta l'approssimazione in cui una molecola interagisce con le molecole vicine fino ad un certo raggio di cutoff $r_C$. Si vuole ora introdurre una correzione in grado di tener conto anche della parte a lungo range del potenziale, ovvero considerare anche l'interazione tra molecole molto distanti tra loro. Un concetto fondamentale usato per il calcolo del potenziale a $r>r_C$ è quello di *"radial distribution function $g(r)$"* che rappresenta la probabilità (relativa) di trovare una particella a distanza $r$ da una particella scelta come origine, rispetto alla probabilità che si avrebbe in un sistema omogeneo e non correlato (come un gas ideale). Si può far vedere che la media di una qualsiasi funzione che dipende da coppie di posizioni (es $V(r_i, r_j)$) si può calcolare in modo semplice a partire da $g(r)$.  
Nel presente caso per calcolare la media del potenziale di Lennard-Jones $V_{LJ}$ e della pressione $P$ per $r>r_C$ si considera in genere $g(r) \sim 1$, da cui si ricava la tail correction:
$$
\langle V \rangle _{Tail} = \dfrac{8}{3} \pi \rho \left[\dfrac{1}{3} \left(\dfrac{\sigma}{r_{cut}} \right)^9 -\left(\dfrac{\sigma}{r_{cut}} \right)^3 \right] \cdot N
$$
$$
\langle P \rangle _{Tail} = \dfrac{32}{3} \pi \rho \left[\dfrac{1}{3} \left(\dfrac{\sigma}{r_{cut}} \right)^9 -\dfrac{1}{2} \left(\dfrac{\sigma}{r_{cut}} \right)^3 \right] \cdot \rho
$$
dove le grandezze fisiche contenute nelle formule sono espresse (come nell'esercizio 4) in termini di unità di Lennard-Jones $\sigma$, $\epsilon$ e $\epsilon/k_B$.

Siccome il grafico non permette di distinguere visivamente quando il sistema è di fatto giunto a equilibrazione, si può dare una stima un po' più quantitativa di quando questo accade, utilizzando la **media mobile**: 
l'idea è considerare il set di $1000$ configurazioni e calcolare la media su un blocco di $100$, facendo poi scorrere il blocco di uno step alla volta. In parole semplici calcolo la media sui primi $100$ valori, poi calcolo la media tra il primo valore e il 101-esimo, etc...

Un ulteriore test è quello della deviazione standard relativa mobile, ovvero come varia la deviazione standard shiftando il blocco di un unità alla volta, rapportata al valore medio.

Per procedere con l'analisi è interessante capire quanto dovrebbero essere grandi i blocchi utilizzati per il data blocking dei risultati. Un primo criterio è dato dall'**autocorrelazione**, ovvero una stima quantitativa di come i dati in una serie sono correlati tra di loro, nel caso in esame misura il grado di dipendenza tra i campionamenti successivi. Si ricordi che quello che si vuole realizzare per ottenere risultati più accurati è una serie di misure con configurazioni il più possibile indipendenti tra di loro.  
La definizione matematica di **autocorrelazione** è la seguente:
$$
AC_{[x]} = \dfrac{ \text{cov}_{[x,x]}(t, t+k) } { \text{cov}_{[x,x]}(t, t) }
$$

con

$$
\text{cov}_{[x,x]}(t, t+k) = \langle x(t)\cdot x(t+k) \rangle - \langle x(t) \rangle \cdot \langle x(t+k) \rangle,
$$

dove $k$ è il cosiddetto "lag", ovvero la distanza temporale (in termini di steps Monte Carlo) tra i due campionamenti e $x$ è la variabile che si sta considerando.

Si può riscrivere la formula dell'autocorrelazione nella seguente formula più operativa per l'applicazione che si sta realizzando:

$$
R_k = \frac{1}{(N - k) \sigma^2} \sum_{i=1}^{N - k} (x_i - \mu)(x_{i+k} - \mu)
$$

In questa formula, $R_k$  rappresenta il valore dell'autocorrelazione al lag $k$. 
$ \; N $ è la lunghezza complessiva della serie temporale. Il simbolo $\mu$ indica la media della serie originale $x$, mentre $\sigma^2$ rappresenta la varianza, sempre calcolata sulla serie originale. I termini $x_i$ e $x_{i+k}$ sono, rispettivamente, il valore della serie al tempo $i$ e il valore al tempo $i+k$, ovvero traslato in avanti di $k$ passi.

Di seguito è riportata l'analisi di come varia l'autocorrelazione della variabile $x = U/N$ in funzione del "lag", prendendo in considerazione lag compresi nell'intervallo $[1, 400]$ step Monte Carlo per i 3 differenti stati della materia (*NB: le medie si intendono su tutto il dataset di $M=5 \times 10^5$*):

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6)) 
plt.title(r"Autocorrelazione", fontsize=18)

UN_solido_tot = np.loadtxt("Solido/potential_VS_timestep_SOLIDO_tot.dat", unpack= True, usecols=(1), skiprows=1) 
UN_liquido_tot = np.loadtxt("Liquido/potential_VS_timestep_LIQUIDO_tot.dat", unpack= True, usecols=(1), skiprows=1) 
UN_gassoso_tot = np.loadtxt("Gassoso/potential_VS_timestep_GAS_tot.dat", unpack= True, usecols=(1), skiprows=1) 

def autocorr (U_N, k):
    x = np.array(U_N)
    n = len(x)
    mu = np.mean(x)
    var = np.var(x)
    return 1. / ((n-k)*var) * np.sum((x[:n-k] - mu) * (x[k:] - mu))

n_lags =400
lag_range = np.arange(1, n_lags+1, 1)
AC_vector = np.zeros(n_lags)


#Per solido
for k in lag_range:
    AC_vector [k-1] = autocorr (UN_solido_tot, k)

plt.plot(lag_range, AC_vector, label = "Solido", color = 'orange')

#Per liquido
AC_vector = np.zeros(n_lags)
for k in lag_range:
    AC_vector [k-1] = autocorr (UN_liquido_tot, k)

plt.plot(lag_range, AC_vector, label = "Liquido", color = 'blue')


#Per gas
AC_vector = np.zeros(n_lags)
for k in lag_range:
    AC_vector [k-1] = autocorr (UN_gassoso_tot, k)

plt.plot(lag_range, AC_vector, label = "Gas", color = 'green')


plt.xlabel('Lag dimension', fontsize = 14)
plt.ylabel('Autocorrelation', fontsize = 14)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.ylim(-0.02, 0.25)

plt.legend(loc='upper right', prop={'size': 12})
plt.grid(True)
plt.show()
```

Dai risultati si evince che i campionamenti decorrelano in maniera diversa a seconda dello stato della materia considerato. In particolare lo stato gassoso, che per sua natura è lo stato più caotico decorrela in effetti dopo un numero molto basso di step (circa $30$), mentre gli altri due stati della materia neccessitano di più steps per mostrare campionamenti decorrelati tra di loro (circa $200$ step per lo stato solido e circa $300$ step per lo stato liquido) . Si dovrebbero dunque prendere blocchi contenenti ciascuno un numero superiore al numero di steps necessario ad osservare dati non correlati tra di loro.

**Data Blocking**

Un altro criterio utilizzato per decidere quanti blocchi realizzare con gli $M=500000$ campionamenti, si basa sull'osservare l'andamento dell'errore statistico in funzione di $L$ (numero di campionamenti per blocco) in un intervallo compreso tra $L=10$ e $L=5000$ per multipli di 10. 

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) 
plt.title(r"Errore al variare della dimensione dei blocchi", fontsize=18)

UN_solido_tot = np.loadtxt("Solido/potential_VS_timestep_SOLIDO_tot.dat", unpack= True, usecols=(1), skiprows=1) 
UN_liquido_tot = np.loadtxt("Liquido/potential_VS_timestep_LIQUIDO_tot.dat", unpack= True, usecols=(1), skiprows=1) 
UN_gassoso_tot = np.loadtxt("Gassoso/potential_VS_timestep_GAS_tot.dat", unpack= True, usecols=(1), skiprows=1) 

M = len(UN_gassoso_tot)

x = np.arange(10, 5010, 10)


def error_calc (UN):
    y = np.array(UN)
    err = np.zeros(len(x))
    j = 0
    for L in x:
        N = int(M/L) #numero di blocchi a seconda di quanto vale L
        ave = np.zeros(N)
        for i in range(N):
            ave[i] = np.mean(y[i*L : i*L+L])
        err [j] =  np.sqrt(np.var(ave)/N)
        j+=1
    return err


#Solido
plt.plot(x, error_calc(UN_solido_tot), label='Solido', color='orange')

#Liquido
plt.plot(x, error_calc(UN_liquido_tot), label='Liquido', color='blue')

#Gas
plt.plot(x, error_calc(UN_gassoso_tot), label='Gas', color='green')
    
plt.xlabel('Dimensione dei blocchi', fontsize = 14)
plt.ylabel('Errore', fontsize = 14)

plt.xticks(fontsize = 7, ticks = np.arange(0, 5010, 200))
plt.yticks(fontsize = 13)

plt.legend(loc='lower right', prop={'size': 12})
plt.grid(True)
plt.show()
```

Si osserva un comportamento di saturazione del valore di errore superata una certa soglia per la dimensione dei blocchi che è pari a circa $100$ campioni per blocco nella fase gassosa, circa $500$ nella fase solida e circa $800$ nella fase liquida. Si nota inoltre che per blocchi troppo piccoli si sottostima l'errore, in quanto si stanno considerando variabili in qualche modo correlate tra di loro (per via della natura pseudocasuale dei campionamenti).  

Per avere decorrelazione tra i campionamenti e al contempo non sottostimare l'errore è opportuno quindi utilizzare blocchi sufficientemente grandi. In quel che segue, per semplicità si è scelto di uniformare i casi solido, liquido e gassoso considerando blocchi di dimensione $1000$.

## Ex 7.03 - Radial distribution function $g(r)$

Si vuole ora integrare un algoritmo per il calcolo della $g(r)$ descritta ad inizio del presente notebook, all'interno del codice "NSL_SIMULATOR" già utilizzato nei punti precedenti e nell'esercitazione 4. 
Ricordando che $g(R)$ è definita come la probabilità (relativa) di trovare una particella a distanza $r$ da una particella scelta come origine, rispetto alla probabilità che si avrebbe in un sistema omogeneo e non correlato (come un gas ideale), pertanto:
$$
g(r) = \frac{\text{probabilità particelle a distanza $r$}}{\text{probabilità distribuzione uniforme}}.
$$
Siccome non è possibile conoscere a priori la probabilità di trovare due particelle ad una specifica distanza $r$ tra loro, l'idea è misurare tale probabilità locale campionando le distanze effettive tra tutte le particelle del sistema. Ciò viene fatto per ogni step Monte Carlo (aka configurazione del sistema) e in seguito si procede con il solito block averaging come per le altre grandezze fisiche considerate. 

Operativamente si prende l'intervallo di valori di $r$ per i quali si vuole calcolare $g(r)$ (nel presente esercizio $r \in [0,L/2]$, dove L è il lato della cella presa in considerazione), e lo si divide in un certo numero di sottointervalli ($n_{bins} = 200$) ciascuno avente lunghezza pari a $dr = \dfrac{L/2} {n_{bins}}$ da considerare come "contenitori di particelle". Si misura poi la distanza reciproca $\tilde{r}$ per ogni coppia di particelle, incrementando di $2$ il contenuto del bin contenente quella coppia, ovvero quel sottointervallo $[r^*, r^* + dr]$ t.c. $\tilde{r} \in [r^*, r^* + dr]$.

Dopo aver considerato tutte le coppie di particelle, bisogna infine normalizzare il contenuto di ogni bin con il fattore di normalizzazione:
$$
f_{norm} = \rho N_{part} \cdot \dfrac{4}{3}\pi \left[\left(r+dr \right)^3 -r^3 \right]
$$
che corrisponde al numero di particelle aspettato in un certo bin in caso di distribuzione uniforme $\rho$.  
Dunque si ottiene una stima di $g(r)$ valutata a tratti.

*Nel seguente sono stati considerati $200$ blocchi da $1000$ campionamenti ciascuno per i motivi esposti nell'analisi precedente*.

I risultati ottenuti tramite Monte Carlo verranno infine confrontati con quelli derivanti da simulazioni di Dinamica Molecolare nel regime microcanonico (NVE), svolte in condizioni termodinamiche analoghe, con particolare attenzione alla forma e al comportamento della funzione $g(r)$.

```python
import numpy as np
import matplotlib.pyplot as plt

####    SOLIDO   ####
P_NVT, P_NVT_err = np.loadtxt("Solido/NVT/pressure.dat", unpack=True, usecols=(2,3), skiprows=1)
U_NVT, U_NVT_err = np.loadtxt("Solido/NVT/potential_energy.dat", unpack=True, usecols=(2,3), skiprows=1)
r, gofr_NVT, gofr_NVT_err = np.loadtxt("Solido/NVT/gofr.dat", unpack=True, usecols=(0,1,2), skiprows=0)
#Dati da simulazione NVE
gofr_NVE, gofr_NVE_err = np.loadtxt("Solido/NVE/gofr.dat", unpack=True, usecols=(1,2), skiprows=0)


x=np.arange(1, len(P_NVT)+1)

plt.figure(figsize=(21, 15))

#Pressione
plt.subplot(2,2,1)

plt.title(r"Pressione", fontsize=18)

plt.errorbar(x, P_NVT, yerr=P_NVT_err, label = "Pressure NVT", color = 'orange')

plt.xlabel('Blocchi', fontsize = 14)
plt.ylabel('Pressure', fontsize = 14)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)


plt.legend(loc='lower right', prop={'size': 12})
plt.grid(True)

#Energia potenziale
plt.subplot(2,2,2)

plt.title(r"Energia potenziale per particella", fontsize=18)

plt.errorbar(x, U_NVT, yerr=U_NVT_err, label = "Potential energy NVT", color = 'orange')

plt.xlabel('Blocchi', fontsize = 14)
plt.ylabel('Potential energy', fontsize = 14)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)


plt.legend(loc='lower right', prop={'size': 12})
plt.grid(True)

#Radial Distr. Func.
plt.subplot(2,2,3)

plt.title(r"$g(r)$", fontsize=18)

plt.errorbar(r, gofr_NVT, yerr=gofr_NVT_err, label = r"$g(r)$ NVT", color = 'orange')
plt.errorbar(r, gofr_NVE, yerr=gofr_NVE_err, label = r"$g(r)$ NVE", color = 'red')


plt.xlabel('r', fontsize = 14)
plt.ylabel(r"$g(r)$", fontsize = 14)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)


plt.legend(loc='upper right', prop={'size': 12})
plt.grid(True)

plt.show()
```

Per quanto riguarda la funzione $g(r)$ si noti come per valori piccoli di $r$ è nulla, ciò è dovuto al carattere repulsivo a breve range del potenziale di tipo Lennard-Jones. Si ha poi un picco su un valore di $r$ intermedio che corrisponde alla distanza tra coppie di particelle più probabile, mentre per valori di $r$ più grandi la funzione oscilla attorno a $1$, che evidenzia la tendenza del sistema di molecole ad assomigliare ad un gas ideale per grandi distanze intermolecolari.

E' inoltre evidente il buon accordo tra i due metodi statistici utilizzati NVE (energia fissata) e NVT (temperatura fissata).
