## Esercizio 2.01 - Calcolo di un integrale 1D con metodo Monte Carlo

L'integrale da calcolare con metodo Monte Carlo è:
$$
I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1
$$

### Metodo 1: Sampling da distribuzione uniforme

Un primo metodo per stimare l'integrale con metodo Monte Carlo è quello di considerare l'integrale 
$$ I = \int_a^b f(x) dx
$$

Dividendo l'intervallo di integrazione in segmenti di pari lunghezza $\Delta x = (b-a)/N$ e mandando $N \to \infty$, si può scrivere l'integrale nella forma:
$$
I = \lim_{N \to \infty} \sum_{i=1}^N \Delta x f(x_i) = (b-a) \lim_{N \to \infty} \frac{1}{N} \sum_{i=1}^N f(x_i)
$$
che infine corrisponde a calcolare:
$$
I = (b-a) \langle f(x) \rangle_{[a,b]}
$$
dove la media di $f(x)$ è calcolata sull'intervallo $[a,b]$.

Si genera dunque un numero pari a $M$ di estrazioni di $x_i$ uniformemente estratti dall'intervallo $[a,b]$ e li si raggruppa in $N$ blocchi ciascuno contenente $L = M/N$ valori. Si calcola poi la media cumulativa di $f(x)$ con relativo errore utilizzando un numero variabile di blocchi, da $1$ a $N$ alla stessa maniera dell'esercitazione 1.  
Si ottiene il seguente risultato:

```python
import matplotlib.pyplot as plt
import numpy as np
ave, err = np.loadtxt("MC_unif.txt", unpack=True, usecols=(0,1), skiprows=1)
```

```python
N=100 #numero di blocchi
x_axis = np.arange (1, N+1)
plt.figure(figsize=(8, 6))
plt.title("Integrale calcolato con Monte Carlo uniforme", fontsize=17, weight="bold")
plt.errorbar(x_axis,ave,yerr=err, label="Simulato")
plt.xlabel(r"# Blocchi", fontsize=15)
plt.ylabel(r"$ \quad \langle \quad \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx \quad \rangle \quad $", fontsize=15)
plt.axhline( y=1, xmin=0, xmax=100, color='red', label="Val. atteso")
plt.legend()
plt.grid(True)
plt.show()

print("La media cumulativa dopo 100 blocchi è: ", ave[-1], "con un errore di: ", round(err[-1],5), ", ovvero entro ", abs(round((ave[-1]-1)/err[-1],2)), "sigma.")
```

### Metodo 2: importance sampling

Nel metodo al punto sopra si è calcolato un'integrale del tipo:
$$
I = \int_0^1 f(x)p(x) dx
$$
con $p(x)= 1$, ovvero estraendo le variabili casuali $x_i$ da una distribuzione uniforme.  
Si vuole ora procedere con un metodo più preciso e rapido nella convergenza al risultato che prevede l'utilizzo di una funzione di probabilità $p(x)$ che approssima la funzione da integrare.  

**Il senso di questo metodo è estrarre più spesso valori di $x$ per cui la funzione $f(x)$ è più grande così da dare più peso ai punti dove la funzione assume valori significativi, e meno peso ai punti in cui il valore della funzione non darebbe un grosso contributo all'integrale.**

Si ricordi che la funzione da integrare era:
$$
f(x) = \frac{\pi}{2} \cos \left(\frac{\pi}{2} x \right)
$$
Si consideri ora lo sviluppo al primo ordine di $f(x)$ intorno a $x=1/2$ (punto medio dell'intervallo di integrazione):  
<small> *Nota: si potrebbe anche sviluppare sull'estremo $x=1$ per semplificare i calcoli, in questo caso il risultato non cambierebbe significativamente* </small>
$$
f(x) = \frac{\pi \sqrt{2}}{4} - \frac{\pi^2 \sqrt{2}}{8} \left( x - \frac{1}{2} \right) + ...
$$
Per essere una densità di probabilità deve essere normalizzata a 1 se integrata tra 0 e 1:
$$
1 \overset{!}{=} N \int_0^1  \frac{\pi \sqrt{2}}{4} - \frac{\pi^2 \sqrt{2}}{8} \left( x - \frac{1}{2} \right)
$$
$$
\Rightarrow \quad N = \frac{4}{\sqrt{2} \pi}
$$
$$
\Rightarrow \quad p(x) = 1 - \frac{\pi}{2} \left( x - \frac{1}{2} \right)
$$
Dunque l'integrale I scritto sopra si scrive come:
$$
I = \int_0^1 \frac{f(x)}{p(x)} p(x) dx 
$$
definendo $g(x) \equiv \frac{f(x)}{p(x)}$ la nuova integranda con $f(x)$ e $p(x)$ definite sopra.  

L'integrale quindi può essere calcolato come:
$$
I \approx \frac{1}{N} \sum_i^N g(x_i)
$$
con $x_i$ estratti dalla distribuzione $p(x)$.  
Per estrarre valori di $x_i$ dalla distribuzione $p(x)$ si utilizza il metodo dell'inversa cumulativa (già incontrato nell'esercitazione 2). Si scrive dunque:
$$
\begin{aligned}
y = F(x) &= \int_0^x p(x') dx' \\
            &= \int_0^x 1 - \frac{\pi}{2} \left( x' - \frac{1}{2} \right) = \\
           &= \left(1 + \frac{\pi}{4} \right)x - \frac{\pi}{4} x^2
\end{aligned}
$$
Invertendo la funzione con alcuni passaggi algebrici si giunge a:
$$
x = \frac{(4 + \pi) \pm \sqrt{(4 + \pi)^2 - 16\pi y}}{2\pi}
$$
dove $x$ è una variabile che segue la distribuzione $p(x)$ se estraggo $y$ uniformemente nell'intervallo $[0,1)$. Inoltre per avere $x=0$ quando $y=0$ scelgo la soluzione con il segno "$-$" , anche per ottenere valori estratti nell'intervallo giusto $[0,1]$.  

Si riporta dapprima l'istogramma che mostra come i valori di $x_i$ estratti seguono effettivamente la distribuzione $p(x)$:

```python
import matplotlib.pyplot as plt
import numpy as np
ran_numbers = np.loadtxt("random_numbers.txt", unpack=True, usecols=(0), skiprows=0)
```

```python
plt.figure(figsize=(6,4))
plt.title("Istogramma delle estrazioni di valori da p(x)", fontsize=18)
plt.xlabel(r'$x_i$')
plt.ylabel("Estrazioni")
plt.hist(ran_numbers, bins=30, density=True) 
plt.show()
```

Nello stesso spirito del punto precedente si riporta di seguito la media cumulativa per l'integrare $I$ al variare dei blocchi.

```python
import matplotlib.pyplot as plt
import numpy as np
ave, err = np.loadtxt("MC_importance.txt", unpack=True, usecols=(0,1), skiprows=1)
```

```python
N=100 #numero di blocchi
x_axis = np.arange (1, N+1)
plt.figure(figsize=(8, 6))
plt.title("Integrale calcolato con Monte Carlo importance sampling", fontsize=17, weight="bold")
plt.errorbar(x_axis,ave,yerr=err, label="Simulato")
plt.xlabel(r"# Blocchi", fontsize=15)
plt.ylabel(r"$ \quad \langle \quad \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx \quad \rangle \quad $", fontsize=15)
plt.axhline( y=1, xmin=0, xmax=100, color='red', label="Val. atteso")
plt.legend()
plt.grid(True)
plt.show()

print("La media cumulativa dopo 100 blocchi è: ", ave[-1], "con un errore di: ", round(err[-1],5), ", ovvero entro ", abs(round((ave[-1]-1)/err[-1],2)), "sigma.")
```

Dal grafico appena mostrato si evince come con questo secondo metodo il livello di confidenza è più basso rispetto al caso del sampling uniforme, ciò è dovuto al fatto che la varianza è molto più bassa (circa 4 volte inferiore). Inoltre guardando meglio la scala dell'asse y nei due grafici si può notare come nel secondo il risultato converga molto più rapidamente al valore atteso.

## Es 2.02 - Random walks

### Random walk discreto
Si procede ora con lo studio dei random walks, si considera cioè un cammino generato casualmente all'interno dello spazio 3D. Partendo dall'origine degli assi, viene eseguito un passo alla volta di lunghezza $1$ in una delle direzioni $x$, $y$ o $z$ sia nel verso positivo che negativo scelta casualmente in maniera uniforme.
Si sono generati quindi $M = 10^4$ differenti random walks da 100 passi ciascuno costruiti in questo modo e si è costruito un vettore che contenesse la distanza dall'origine (al quadrato) all'i-esimo passo avente la forma:
$$moves = \left( \quad r_1^2(1), r_1^2(2), \quad ... \quad, r_1^2(100), r_2^2(1), \quad ... \quad, r_2^2(100), \quad ... \quad, r_{10000}^2(100) \quad \right)
$$
dove il pedice indica il j-esimo random walk ($j : 1 \rightarrow 10000$), mentre il numero tra parentesi indica l'i-esimo passo ($i : 1 \rightarrow 100$).  
Per ogni possibile numero di passi ($i : 1 \rightarrow 100$) si sono poi raggruppati i random walks in $B = 100$ blocchi da $L = 100$ simulazioni ciascuno, costruendo cosi una matrice $M$ delle medie del tipo:
$$
\left(
\begin{array}{cccc}
\langle r^2_{1}(1) \rangle & \langle r^2_{2}(1) \rangle & ... & \langle r^2_{B}(1) \rangle \\ 
\langle r^2_{1}(2) \rangle & \langle r^2_{2}(2) \rangle & \ddots & \langle r^2_{B}(2) \rangle \\
\vdots & \vdots & \vdots & \vdots  \\
\langle r^2_{1}(100) \rangle & \langle r^2_{2}(100) \rangle  & ... &  \langle r^2_{B}(100) \rangle
\end{array}
\right)
$$

dove il pedice rappresenta l' s-esimo blocco di 100 random walks per cui il valore di $r^2$ è stato mediato, mentre ogni colonna (il numero tra parentesi) indica di quanti passi sto considerando il random walk.  
Il prossimo passaggio è mediare sulle colonne e calcolare la deviazione standard, in questo modo il risultato finale risulta essere un vettore delle medie delle $r_N^2$ posizioni finali al variare del numero di passi N. (NB: siccome voglio plottare $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ e non $\langle |\vec{r}_N|^2 \rangle_{RW}$, anche l'errore deve essere ricalcolato sulla radice utilizzando la propagazione degli errori).

Nel grafico che segue viene plottato quest'ultimo vettore e fittato con una soluzione del tipo $y = k \sqrt{N}$ che testa il carattere diffusivo del risultato ottenuto.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats import chi2

ave, err = np.loadtxt("RW_discreto.txt", unpack=True, usecols=(0,1), skiprows=1)
```

```python
N=100 #max numero di passi
x_axis = np.arange (1, N+1)

#Funzione con cui fittare
def func (x, k):
    return k * np.sqrt(x)

#Fitto la funzione: func è la funzione con cui fittare dipendente da x e da 1 o più parametri, x_axis e ave sono 
#rispettivamente la variabile indipendente e la variabile dipendente prese dai dati.
k_opti, _ = curve_fit(func, x_axis, ave)
k_val = k_opti[0]  #perchè la funzione curve_fit restituisce un array di parametri

##Voglio ora calcolare il chi_quadro
#Calcolo i valori di y aspettati in base alla funzione fit
y_expected = func (x_axis, k_val)     
#Calcolo il chi_quadro usando  (togliendo il primo dato che ha varianza nulla)
chi2_pears = np.sum(((ave[1:]-y_expected[1:])/err[1:])**2)           
# Chi-quadro ridotto (divido il chi2 per i GDL=N-#parametri_fittati = 99-1 = 98)
GDL = 98.0
chi2_red = chi2_pears / GDL
# p-value
p_value = 1 - chi2.cdf(chi2_pears, GDL)
# Percentuale di compatibilità (100% = fit perfetto)
compatibility = p_value * 100
plt.figure(figsize=(8, 6))
plt.title(r'Distanza raggiunta in media dal RW al variare di N', fontsize=17, weight="bold")
plt.plot(x_axis, func(x_axis, k_opti), 'ro', label=rf'Fit con $f = {k_val:.3f}\sqrt{{N}}$')
plt.errorbar(x_axis,ave,yerr=err, label="Simulato")
plt.xlabel(r'N passi', fontsize=15)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$', fontsize=15)
plt.legend()
plt.grid(True)
plt.show()

print("La posizione media dopo 100 passi è: ", ave[-1], "+-", round(err[-1],5) ,"\n")
print("Il chi_quadro ridotto è: ", round(chi2_red,3), "\n")
print("La compatibilità con la funzione di fit è del ", round(compatibility,4), " %")
```

Dal fit appena mostrato si osserva che la soluzione trovata ha una compatibilità molto alta con una funzione della forma $y = k \sqrt{N}$, tale soluzione rappresenta un modo classico browniano in cui $N$ è da intendersi come il tempo, mentre $k$ è il coefficiente di diffusione.

### Random walk continuo
Si vuole ora analizzare il caso in cui ad ogni passo può essere presa una qualsiasi direzione anzichè sceglierne una appartenente ad un set discreto (aka nelle direzioni dei 3 assi) come fatto nel caso precedente.  
Procedendo in coordinate sferiche si potrebbe pensare che basti estrarre uniformemente $\phi \in [0,2\pi)$ e $\theta \in [0,\pi)$, ma cosi facendo non si stanno veramente considerando tutte le direzioni come equiprobabili.  
Si può notare questo fatto plottando la distribuzione di punti sulla superficie di una sfera unitaria estratti in questa maniera errata: il risultato è che i punti sono più concentrati sui poli.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Numero di punti da generare
N = 10000

# Estrazione casuale:
# theta: angolo polare in [0, pi]
# phi: angolo azimutale in [0, 2pi]
theta = np.random.uniform(0, np.pi, N)
phi = np.random.uniform(0, 2 * np.pi, N)

# Conversione in coordinate cartesiane
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

# Creazione del doppio plot
fig, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'projection': '3d'})

#############PRIMO PLOT#################
# Creazione della figura 3D
axes[0].scatter(x, y, z, s=1, c='blue', alpha=0.3)

# Impostazione degli assi e del titolo
axes[0].set_title("Vista dall'alto")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")

# Orientazione dall'alto
axes[0].view_init(elev=90, azim=0)
axes[0].set_xticks([])
axes[0].set_yticks([])
axes[0].set_zticks([])

# Mantiene le proporzioni corrette
axes[0].set_box_aspect([3, 3, 3])

#############SECONDO PLOT#################
# Creazione della figura 3D
axes[1].scatter(x, y, z, s=1, c='blue', alpha=0.3)

# Impostazione degli assi e del titolo
axes[1].set_title("Vista frontale")
axes[1].set_ylabel("Y")
axes[1].set_zlabel("Z")

# Orientazione frontale: elevazione 0 e azimut 0
axes[1].view_init(elev=0, azim=0)
axes[1].set_xticks([])
axes[1].set_yticks([])
axes[1].set_zticks([])

# Mantiene le proporzioni corrette
axes[1].set_box_aspect([3, 3, 3])

plt.show()
```

La motivazione è la seguente ed ha carattere puramente geometrico:  
l'elemento infinitesimo di superficie in coordinate sferiche per una sfera unitaria è
$$
dA = \sin{\theta} \, d\theta \, d\phi
$$
dunque la densità superficiale di punti posto che la distribuzione sia uniforme in theta (aka indipendente da theta, $N(\theta) = N$ ) sarebbe
$$
\rho(\theta) = \frac{dN(\theta)}{dA} = \frac{dN}{\sin{\theta} \, d\theta \, d\phi} .
$$
Manifestamente è una densità che diverge in $\theta = 0, \pi$, motivando così la maggior concentrazione di punti ai poli.  
Per annullare questo effetto basta prendere una distribuzione non uniforme in $\theta$ che segua l'andamento $\sin{\theta}$.

Utilizzando il metodo dell'inversa cumulativa per estrarre punti dalla distribuzione $p(x) = \sin{x}$ si scopre che i valori di $\theta$ vanno estratti da:
$$
x = \cos^{-1}(1 - 2U)
$$
dove $U \in [0, 1)$ estratto uniformemente.  

Il resto dell'analisi è svolta come nel caso discreto, portando ai seguenti risultati:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats import chi2

ave, err = np.loadtxt("RW_continuo.txt", unpack=True, usecols=(0,1), skiprows=1)
```

```python
N=100 #max numero di passi
x_axis = np.arange (1, N+1)

#Funzione con cui fittare
def func (x, k):
    return k * np.sqrt(x)

#Fitto la funzione: func è la funzione con cui fittare dipendente da x e da 1 o più parametri, x_axis e ave sono 
#rispettivamente la variabile indipendente e la variabile dipendente prese dai dati.
k_opti, _ = curve_fit(func, x_axis, ave)
k_val = k_opti[0]  #perchè la funzione curve_fit restituisce un array di parametri

##Voglio ora calcolare il chi_quadro
#Calcolo i valori di y aspettati in base alla funzione fit
y_expected = func (x_axis, k_val)     
#Calcolo il chi_quadro usando  (togliendo il primo dato che ha varianza nulla)
chi2_pears = np.sum(((ave[1:]-y_expected[1:])/err[1:])**2)           
# Chi-quadro ridotto (divido il chi2 per i GDL=N-#parametri_fittati = 99-1 = 98)
GDL = 98.0
chi2_red = chi2_pears / GDL
# p-value
p_value = 1 - chi2.cdf(chi2_pears, GDL)
# Percentuale di compatibilità (100% = fit perfetto)
compatibility = p_value * 100
plt.figure(figsize=(8, 6))
plt.title(r'Distanza raggiunta in media dal RW al variare di N', fontsize=17, weight="bold")
plt.plot(x_axis, func(x_axis, k_opti), 'ro', label=rf'Fit con $f = {k_val:.3f}\sqrt{{N}}$')
plt.errorbar(x_axis,ave,yerr=err, label="Simulato")
plt.xlabel(r'N passi', fontsize=15)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$', fontsize=15)
plt.legend()
plt.grid(True)
plt.show()

print("La posizione media dopo 100 passi è: ", ave[-1], "+-", round(err[-1],5) ,"\n")
print("Il chi_quadro ridotto è: ", round(chi2_red,3), "\n")
print("La compatibilità con la funzione di fit è del ", round(compatibility,4), " %")
```

Dal fit appena mostrato si osserva che la soluzione trovata ha una compatibilità molto alta con una funzione della forma $y = k \sqrt{N}$, tale soluzione rappresenta un modo classico browniano in cui $N$ è da intendersi come il tempo, mentre $k$ è il coefficiente di diffusione.
